```python
import pandas as pd
df = pd.read_csv('2014-2018_Road_Crash.csv', sep = ',')
```

```python
df = df[["LGA_Name", "Year", "Total_Cas", "REPORT_ID"]]
df
```

```python
df1 = df.drop("Total_Cas", 1)
df1
```

```python
df1 = df1.groupby(["LGA_Name", "Year"]).count().reset_index()
df1
```

```python
df2 = df.drop("REPORT_ID", 1)
df2
```

```python
df2 = df2.groupby(["LGA_Name", "Year"])["Total_Cas"].sum().reset_index()
```

```python
df2
```

```python
df3 = pd.concat([df1, df2], axis = 1, join='inner').T.drop_duplicates().T
df3 = df3.rename(columns={'REPORT_ID': 'Total_Crash'})
df3.reset_index()
df3
```

```python
df3.to_csv("2014-2018_Crash_Cas.csv", index=False)
```
